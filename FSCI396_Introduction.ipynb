## How do Choose specific data

```python
# Typing out 'df.columns' can help you see the exact naming of each column. 

df.columns
```

```python
#If you want to print a single column, you can do this by following: df['ColumnName']

courses = df['Course']

print(courses)
```

```python
# You may also determine how many unique answers there are in a given column using .unique() 

UniqueCourses = df['Course'].unique()

print(UniqueCourses)
```

```python
# If you want to choose a column with only a specific answer, you can do this by following: 
# df[df['ColumnName'==''Answer]]

# For example, if you want all data from a MATH2XX, write:
Math = df[df['Course']=='MATH2XX']
Math
```

```python
# How to call multiple courses at once 
MathAndPhysics = df[(df['Course']=='MATH2XX') |  (df['Course']=='PHYS1XX') ]

MathAndPhysics
```

```python
 # Alternatively, if you want to find responses who are in Math AND female, you would use the & symbol: 
# Notice, that 'Woman ' has a space at the end and needs to be included. 
# Sometimes Microsoft forms adds random symbols/spaces that humans might not notice, but computers will notice
# This is why using .unique() can be useful to avoid these issues!

MathAndWoman = df[(df['Course']=='MATH2XX') &  (df['Gender']=='Woman ') ]

MathAndWoman
```

## Explore the Data Visually 

```python
# Plot some of this!

# In the code here, you can customize this plot by changing the below lines
title = 'Type Title Here' # Changes the the title on the figure 
xlabel = 'xlabel' # Changes the x axis label
ylabel = 'ylabel' # Changes teh y axis label
filename = 'filename' # Saves the figure to this file name


# We remove state anxiety because is has a different min/max values
fields = ['Rehearsal','Elaboration','SelfEfficacy','Anxiety_test','Anxiety_writing','Anxiety_speaking','Anxiety_group']

df[fields].plot.density()

plt.title(title)
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.legend()

# The below line is commented out for running in the shared folder. 
# uncomment it to save the figure when you've copied the script to your personal folder/directory. 
# plt.savefig(filename)
```

```python
# We can also look and compare multiple columns in one go 

# Compare distributions of multiple columns
columns = ['Rehearsal','Elaboration']    # List of columns to compare
scalemax = 7               # Maximum score for histogram bins (matches scale of survey)

# pairplot shows scatterplots off-diagonal and distributions on the diagonal
Distrib = sns.pairplot(df[columns], diag_kws={'bins': scalemax})
Distrib.fig.suptitle(f"Pairplot for {columns}", y=1.03)
plt.show()

# How to read the figure:
    # Diagonal plots: distribution of each column (histogram)
        # For example, the top left shows the distribution of values for rehearsal 
        # Note, the scales are normalized; the above is easier to interpret
    # Off diagonal plots: scatterplots showing relationships between listed columns 
        # For example, the bottom left scatter plot shows the relation between rehearsal and elaboration
```
